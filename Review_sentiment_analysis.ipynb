```python
import pandas as pd
import re
import nltk
import spacy
from spacy.lang.en import English
nlp = spacy.load('en_core_web_md')
```

```python
Review1 = open('Review1.txt','r').read()
Review2 = open('Review2.txt','r').read()
Review3 = open('Review3.txt','r').read()
```

```python
print(Review1)
```

```python
def clean(text):
    text = re.sub(r'[^a-zA-Z0-9]',' ',text)
    text = re.sub('\d+',' ',text)
    text = re.sub('\s+',' ',text)
    text = text.lower()
    return text
```

```python
Clean_Review1 = clean(Review1)
```

# Manual method

```python
positive = ['Excellent','Great','Wonderful','Marvelous','Phenomenal','Amazing','Fantastic','Impressive','Outstanding','Stupendous','Terrific','Superb','Incredible','Fabulous','Perfect','Phenomenal','Impressive','Joyful','Glowing','Radiant']
Negative = ['Terrible','Awful','Dreadful','Horrible','Unpleasant','Miserable','Disgusting','Abominable','Abusive','Appalling','Atrocious','Deplorable','Detestable','Loathsome','Nasty','Offensive','Repulsive','Vile','Worthless','Nauseating']
Neutral = ['Adequate','Appropriate','Decent','Fine','Good','Okay','Average','Normal','Common','Regular','Ordinary','Average','Normal','Usual','Standard','Routine','Everyday','Fair']
```

```python
%time

pos_c =0
neg_c =0
neu_c =0
others_c =0

words = nltk.word_tokenize(CleanReview3)

for token in words:
    if token in positive:
        pos_c+=1
    elif token in Negative:
        neg_c+=1
    elif token in Neutral:
        neu_c+=1
    else:
        others_c+=1
print('positive count',pos_c)
print('negative count',neg_c)
print('neutral count',neu_c)
print('others count',others_c)
```

# Sentiment Analysis tools from NLTK library

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()
sentiment_scores1 = sia.polarity_scores(Clean_Review1)

print("Sentiment scores: ", sentiment_scores1)


positive_words = 0
negative_words = 0
neutral_words = 0


tokens = nltk.word_tokenize(Clean_Review1)

for word in tokens:
    sentiment_score = sia.polarity_scores(word)['compound']
    if sentiment_score > 0:
        positive_words += 1
    elif sentiment_score < 0:
        negative_words += 1
    else:
        neutral_words += 1


print("Positive words: ", positive_words)
print("Negative words: ", negative_words)
print("Neutral words: ", neutral_words)
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

lemmatized_strings = [lemmatizer.lemmatize(string) for string in filtered_review1]
print(lemmatized_strings)
```

```python
print('Length of Review1 after Lemmatization',len(lemmatized_strings))
```

# POS Tagging

```python
tagged_strings = nltk.pos_tag(lemmatized_strings)
print(tagged_strings,end='\t')
```
