```python
X=house_df_copy[['Trooms','Nbedrooms','Nbwashrooms','Twashrooms','Nfloors','Expected price']]
Y=house_df_copy['Grade']
```

```python
from sklearn.model_selection import train_test_split,cross_validate
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
```

```python
def clasif(est):
    acc=[]
    f1=[]
    precision=[]
    recall=[]
    roc_auc=[]
    r2=[]
    scoring=['accuracy','f1_weighted','precision_weighted','recall_weighted','roc_auc_ovo_weighted','r2']
    cv=cross_validate(est,X,Y,cv=10,scoring=scoring,n_jobs=-1)
    print(cv)
    print('--------------------------------------------------------------------------------------------------------------')
    acc.append(np.mean(cv['test_accuracy']))
    f1.append(np.mean(cv['test_f1_weighted']))
    precision.append(np.mean(cv['test_precision_weighted']))
    recall.append(np.mean(cv['test_recall_weighted']))
    roc_auc.append(np.mean(cv['test_roc_auc_ovo_weighted']))
    r2.append(np.mean(cv['test_r2']))
    
    print('accuracy',acc,'\n','f1_weighted',f1,'\n','precision',precision,'\n','recall',recall,'\n','roc_auc',roc_auc,'\n','r2',r2)
    
```

```python
clasif(GaussianNB())
```

```python
clasif(DecisionTreeClassifier(max_depth=8,random_state=42))
```

```python
clasif(RandomForestClassifier(n_estimators=50,random_state=42,max_depth=8,oob_score=True,n_jobs=-1))
```

```python
base=[]
base.extend([('nb',GaussianNB()),('dt',DecisionTreeClassifier()),('rf',RandomForestClassifier()),('adboost',AdaBoostClassifier())])
```

```python
base
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
clasif(StackingClassifier(estimators=base,final_estimator=LogisticRegression(),cv=10).fit(X,Y))
```

# Interpretation
From the comparison of above Classification models, we choose RandomForest classifier as the best model to solve our buisness problem based on various performance metrics such as accuracy,f1,recall,precision,roc_auc_score and r2 score

```python
!pip install lightgbm
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
# Convert the data into LightGBM format
d_train = lgb.Dataset(x_train, label=y_train)
d_train
```

```python
# Set model hyperparameters
params = {'objective': 'binary',
          'metric': 'binary_logloss',
          'boosting_type': 'gbdt',
          'num_leaves': 31,
          'learning_rate': 0.05,
          'feature_fraction': 0.9,
          'bagging_fraction': 0.8,
          'bagging_freq': 5,
          'verbose': 0}
```

```python
# Train the model
model = lgb.train(params, d_train, num_boost_round=100)
```

```python
# Make predictions on the test set
y_pred = model.predict(x_test)
y_pred
```

```python
# Convert predictions to binary labels
y_pred = np.round(y_pred)
y_pred
```

```python
#lgbm cross-validation
```
